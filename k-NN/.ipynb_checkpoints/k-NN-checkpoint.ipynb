```python
import numpy as np
import operator
import matplotlib.pyplot as plt
```

```python
#k-NN模型
def classify(x, dataSet,labels,k):
    num_dataSet=dataSet.shape[0]
    diffMat=np.tile(x,(num_dataSet,1))-dataSet
    sqMat=diffMat**2
    sqDistances=sqMat.sum(axis=1)
    distances=sqDistances**0.5
    sortedDistIndices=distances.argsort()
    classCount={}
    for i in range(k):
        label=labels[sortedDistIndices[i]]
        classCount[label]=classCount.get(label,0)+1
    sortedClassCount=sorted(classCount.items(),key=operator.itemgetter(1),reverse=True)
    return sortedClassCount[0][0]
```

```python
#读取数据
def file2dataSet(filename):
    file = open(filename)
    file=file.readlines()
    dataSet=np.zeros((len(file),3))
    labelVector=[]
    index=0
    for line in file:
        line=line.strip()
        line=line.split('\t')
        dataSet[index,:]=line[0:3]
        labelVector.append(int(line[-1]))
        index+=1
    return dataSet,labelVector
```

```python
#数据归一化
def autoNorm(dataSet):
    minVals=dataSet.min(0)
    maxVals=dataSet.max(0)
    range=maxVals-minVals
    num_dataSet=dataSet.shape[0]
    normDataSet=dataSet-np.tile(minVals,(num_dataSet,1))
    normDataSet=normDataSet/np.tile(range,(num_dataSet,1))
    return normDataSet
```

```python
#画数据散点图
def showNormDataSet(filename):
    dataSet,labels=file2dataSet(filename)
    normDataSet=autoNorm(dataSet)
    fig=plt.figure()
    pic=fig.add_subplot(111)
    pic.scatter(normDataSet[:,1],normDataSet[:,2],s=12*np.array(labels),c=100*np.array(labels))
```

```python
#测试k-NN模型错误率
def test(filename,k,testPercent):
    dataSet,labels=file2dataSet(filename)
    normDataSet=autoNorm(dataSet)
    num_dataSet=normDataSet.shape[0]
    num_testData=int(num_dataSet*testPercent)
    errorCount=0
    for i in range(num_testData):
        predict=classify(normDataSet[i,:],normDataSet[num_testData:num_dataSet,:],labels[num_testData:num_dataSet],k)
        #print('predict:%d,true:%d'%(predict,labels[i]))
        if(predict!=labels[i]):errorCount+=1
    print('error rate:%f'%(errorCount/float(num_testData)))
```

```python
filename='datingTestSet2.txt'
testPercent=0.1
k=4
showNormDataSet(filename)
test(filename,k,testPercent)
```
